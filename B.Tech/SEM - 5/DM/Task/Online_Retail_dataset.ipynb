## line plot


```python
col1 = df.select_dtypes(include='number').columns[0]
col2 = df.select_dtypes(include='number').columns[1] if df.shape[1] > 1 else col1

plt.figure(figsize=(10, 4))
plt.plot(df[col1], label='Line')
plt.title("Line Plot")
plt.xlabel(col1)
plt.ylabel("Values")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## Histogram

```python
plt.hist(df[col1], bins=20, color='skyblue')
plt.title("Histogram of " + col1)
plt.xlabel(col1)
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

## Scatter

```python
plt.scatter(df[col1], df[col2], c='red')
plt.title("Scatter Plot")
plt.xlabel(col1)
plt.ylabel(col2)
plt.grid(True)
plt.show()
```

## Bar chart (if categorical)

```python
if df[col1].nunique() <= 20:
    df[col1].value_counts().plot(kind='bar', color='orange')
    plt.title("Bar Chart of " + col1)
    plt.grid(True)
    plt.show()
```

## Boxplot

```python
plt.boxplot(df[col1])
plt.title("Boxplot of " + col1)
plt.grid(True)
plt.show()
```

## Pie chart (if categorical)

```python
if df[col1].nunique() <= 6:
    df[col1].value_counts().plot.pie(autopct='%1.1f%%')
    plt.title("Pie Chart")
    plt.ylabel("")
    plt.tight_layout()
    plt.show()
```
