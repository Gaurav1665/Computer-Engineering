### 04) WAP to demonstrate the use of Pie chart.

```python
x = [random.randint(1,50) for i in range(20)]
y = [random.randint(1,50) for i in range(20)]
c = ["r","b"]

plt.pie(y, labels=x, colors=c,  radius=1.5)
plt.pie([1], colors="w",  radius=1)

plt.show()
```

### 05) WAP to demonstrate the use of Bar chart.

### 07) WAP to demonstrate the use of Histogram.

### 08) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
x = [random.randint(1,20) for i in range(10)]
y = [random.randint(1,20) for i in range(10)]

bar = plt.bar(x, y,)
for i in bar:
    height = i.get_height()
    plt.text(i.get_x(), height, f"{height}") 

plt.show()
```

### 09) WAP create a Scatter Plot with several colors in Matplotlib?

```python
x = [random.randint(1,20) for i in range(100)]
y = [random.randint(1,20) for i in range(100)]
z = [random.random() for i in range(100)]

plt.scatter(x, y, c=z)
plt.colorbar()
plt.show()
```
